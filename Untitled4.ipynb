```python
import pandas as pd
import sklearn
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
from sklearn.model_selection import train_test_split
import xgboost
from xgboost import XGBClassifier, XGBRegressor, plot_importance
from sklearn.metrics import classification_report, accuracy_score
```

```python
pd.set_option('display.max_columns',None)
data = pd.read_csv('brooklyn_sales_map.csv', sep = ',')
data[:3]
```

```python
data.isnull().sum() / len(data) * 100
```

```python
data = data.drop('Unnamed: 0', axis = 1)
```

```python
data.head()
```

```python
# data.dropna()
```

```python
data["neighborhood"] = le.fit_transform(data["neighborhood"].astype(str))
```

```python
data.head()
```

```python
quant = data['sale_price'].quantile(0.9)
quant_low = data['sale_price'].quantile(0.1)
```

```python
data = data[data['sale_price'] <quant]
data = data[data['sale_price'] > quant_low]
```

```python
data.shape
```

```python
x = data.drop('building_class_category', axis = 1)
y = data['building_class_category']
```

```python
x_train, x_test, y_train, y_test =  train_test_split(x, y, test_size = 0.3, random_state = 42)
```

```python
xgb = XGBClassifier(n_estimators = 100, learning_rate = 0.3, n_jobs =-1, random_state = 1)
```

```python
xgb.fit(x_train, y_train)
```

```python
y_pred = xgb.predict(x_test)
```

```python
print(classification_report(y_test, y_pred))
```

```python
print(accuracy_score(y_pred, y_test))
```

```python
plot_importance(xgb)
```
